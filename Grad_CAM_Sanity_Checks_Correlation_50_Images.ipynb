```
import pickle

files = open("/content/drive/My Drive/Interpretability Research/Salieny_maps/original_output_binaries.pkl", "rb")
original_output_binaries = pickle.load(files)

files = open("/content/drive/My Drive/Interpretability Research/Salieny_maps/cascade_random_output_binaries.pkl", "rb")
cascade_random_output_binaries = pickle.load(files)

files = open("/content/drive/My Drive/Interpretability Research/Salieny_maps/independent_random_output_binaries.pkl", "rb")
independent_random_output_binaries = pickle.load(files)
```

```
import scipy
import numpy as np
from scipy import integrate
```

```
import scipy.stats as stats
```

```
correlation_cascade_random_list = [1]

for key in cascade_random_output_binaries.keys():

  corr_sum = 0

  for original_output_binary, other_output_binary in zip(original_output_binaries, cascade_random_output_binaries[key]):

    a = original_output_binary.flatten()       

    b = other_output_binary.flatten()  
    
    if np.unique(b).shape[0] == 1:
      corr = 0

    else:
      corr, _ = stats.spearmanr(a, b)

    print(f'corr --> {corr}')
    
    corr_sum += corr

  print(f'corr_sum --> {corr_sum}')
  correlation_cascade_random_list.append(corr_sum/len(cascade_random_output_binaries.keys()))
```

```
import matplotlib.pyplot as plt


x_axis = [i for i in range(1, 18)]
y_axis = correlation_cascade_random_list

plt.figure()
plt.title('Correlation Cascade Randomization Model')
plt.xlabel('''Last Layers : '1' in x_axis is for original one''')
plt.ylabel('Correlation')
plt.plot(x_axis, y_axis)
```

```
correlation_independent_random_list = [1]

for key in independent_random_output_binaries.keys():

  corr_sum = 0

  for original_output_binary, other_output_binary in zip(original_output_binaries, independent_random_output_binaries[key]):

    a = original_output_binary.flatten()       

    b = other_output_binary.flatten()  
    
    if np.unique(b).shape[0] == 1:
      corr = 0

    else:
      corr, _ = stats.spearmanr(a, b)

    print(f'corr --> {corr}')
    
    corr_sum += corr

  print(f'corr_sum --> {corr_sum}')
  correlation_independent_random_list.append(corr_sum/len(independent_random_output_binaries.keys()))
```

```
import matplotlib.pyplot as plt


x_axis = [i for i in range(1, 18)]
y_axis = correlation_independent_random_list

plt.figure()
plt.title('Correlation Independent Randomization Model')
plt.xlabel('''Last Layers : '1' in x_axis is for original one''')
plt.ylabel('Correlation')
plt.plot(x_axis, y_axis)
```
